```python
X  =[] 
Y = []
for category in os.listdir('C:/Users/b51gh/Desktop/ML and DS/38. Project. Text Classification/20_newsgroups'):
    for document in os.listdir('C:/Users/b51gh/Desktop/ML and DS/38. Project. Text Classification/20_newsgroups/'+category):
        with open('C:/Users/b51gh/Desktop/ML and DS/38. Project. Text Classification/20_newsgroups/'+category+'/'+document, "r") as f:
            X.append((document,f.read()))
            Y.append(category)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=1)
```

```python
#dictionary containing frequency of each word occurring in all documents
train_data={}
for i in range(len(X_train)):
    doc_text=X_train[i][1].lower()
    words=re.split(r'\W+',doc_text)
    for j in words:
        if (j in stop_word) or len(j)<=2 or not(j.isalpha()):
            continue
        if j in train_data:
            train_data[j]+=1
        else:
            train_data[j]=1
```

```python
sorted_dic = sorted(train_data.items(), key=operator.itemgetter(1),reverse=True)
sorted_dic
```

```python
import matplotlib.pyplot as plt
x=[]
y=[]
for i in sorted_dic:
    x.append(i[0])
    y.append(i[1])
#plt.plot(x,y)
#plt.show()
```

```python
#This is the list of words or vocabulary we will use
vocabulary=[x[i] for i in range(2000)]
y=[y[i] for i in range(2000)]
vocabulary
```

```python
#x_train_dataset: each row represents a document and the values represent the frequency of all words in our vocabulary
#that are present in the document
x_train_dataset=np.zeros([len(X_train),len(vocabulary)],int)
for i in range(len(X_train)):
    words=X_train[i][1].lower()
    word=re.split(r'\W+',words)
    for j in word:
        if j in vocabulary:
            x_train_dataset[i][vocabulary.index(j)]+=1
```

```python
#x_test_dataset: each row represents a document and the values represent the frequency of all words in our vocabulary
#that are present in the document
x_test_dataset=np.zeros([len(X_test),len(vocabulary)],int)
for i in range(len(X_test)):
    words=X_test[i][1].lower()
    word=re.split(r'\W+',words)
    for j in word:
        if j in vocabulary:
            x_test_dataset[i][vocabulary.index(j)]+=1
x_test_dataset
```

```python
#This dictionary contains frequency of words in vocabulary corresponding to each class and the total documents in that class
dict1={}
total_words=0
Y=np.array(Y_train)
dict1["total_doc"]=len(Y_train)
for i in set(Y_train):
    temp=0
    x_train_curr=x_train_dataset[Y==i]
    dict1[i]={}
    for feature in vocabulary:
        words_count=(x_train_curr[:,vocabulary.index(feature)]).sum()
        dict1[i][feature]=words_count
        temp+=words_count
    dict1[i]["total_count"]=temp
    dict1[i]["total_docs"]=len(x_train_curr)
```

```python
dict1
```

```python
#finds the probability that a document belongs to the current class
def probability(dictionary, x, current_class):
    output = np.log(dictionary[current_class]['total_docs']) - np.log(dictionary['total_doc'])
    words = list(dictionary[current_class].keys())
    for j in range(len(words)-2):
        xj = x[j]
        if xj==0:
            current_xj_probablity=0
        else:
            num=dictionary[current_class][words[j]]+1
            den=dictionary[current_class]['total_count']+len(dictionary[current_class].keys())-2
            current_xj_probablity=np.log(num)-np.log(den)
            output = output + current_xj_probablity
    return output
probability(dict1,x_test_dataset[0],'sci.space')
```

```python
#finds the probability and class prediction for a document and returns the best class acc to best probability
def predictSinglePoint(dictionary, x):
    classes=list(dictionary.keys())
    best_prob=0
    first_run=True
    for i in classes:
        if i in ['total_count','total_doc']:
            continue
        prob=probability(dictionary,x,i)
        if first_run or (prob>best_prob):
            best_prob=prob
            first_run=False
            best_class=i      
    return best_class
predictSinglePoint(dict1,x_test_dataset[0])    
```

```python
#stores the value of predicted class in a list called y_pred and returns it
def predict(dictionary, X_test):
    y_pred = []
    for x in X_test:
        x_class = predictSinglePoint(dictionary, x)
        y_pred.append(x_class)
    return y_pred
```

```python
y_pred_me=predict(dict1,x_test_dataset)
```

```python
#Implementing Text Classification using Multinomial Naive Bayes from sklearn
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
clf=MultinomialNB()
clf.fit(x_train_dataset,Y_train)
y_pred=clf.predict(x_test_dataset)
print("Score on training data:",clf.score(x_train_dataset,Y_train))
print("Score on testing data:",clf.score(x_test_dataset,Y_test))
print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```
